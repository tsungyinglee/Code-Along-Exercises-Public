```python
# Data Science in VS Code tutorial
# https://code.visualstudio.com/docs/datascience/data-science-tutorial

# %pip install pandas numpy matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import requests

url = "https://hbiostat.org/data/repo/titanic3.csv"
response = requests.get(url)
with open("/Users/tsungyinglee/Documents/CodeSpaces/pydata-book-exercises-tyl/data/titanic3.csv", "wb") as f:
    f.write(response.content)

data = pd.read_csv('/Users/tsungyinglee/Documents/CodeSpaces/pydata-book-exercises-tyl/data/titanic3.csv')
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(data):
    # Recode 'sex' column to numeric values
    data['sex'] = data['sex'].map({'male': 1, 'female': 0})
    return data

data_clean = clean_data(data.copy())
data_clean.head()
```

```python
# %pip install seaborn

import seaborn as sns
import matplotlib.pyplot as plt

fig, axs = plt.subplots(ncols=5, figsize=(30,5))
sns.violinplot(x="survived", y="age", hue="sex", data=data_clean, ax=axs[0])
sns.pointplot(x="sibsp", y="survived", hue="sex", data=data_clean, ax=axs[1])
sns.pointplot(x="parch", y="survived", hue="sex", data=data_clean, ax=axs[2])
sns.pointplot(x="pclass", y="survived", hue="sex", data=data_clean, ax=axs[3])
sns.violinplot(x="survived", y="fare", hue="sex", data=data_clean, ax=axs[4])

```
