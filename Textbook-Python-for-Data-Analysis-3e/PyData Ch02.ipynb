```python
# for loops
# Iterate through the sequence and sum the values, skipping None values
sequence = [1, 2, None, 4, None, 5]
total = 0
for value in sequence:
    if value is None:
        continue
    total += value
total
```

```python
# This code iterates through the 'sequence' list and sums the values until it encounters the value 5.
# When the value 5 is found, the loop breaks and stops adding further values.
# The result, 'total_until_5', contains the sum of all values before the first occurrence of 5.
sequence = [1, 2, 0, 4, 6, 5, 2, 1]
total_until_5 = 0
for value in sequence:
    if value == 5:
        break
    total_until_5 += value
total_until_5
```

```python
# This code demonstrates nested loops with a break statement.
# For each value of i from 0 to 3, j also goes from 0 to 3.
# When j becomes greater than i, the inner loop breaks.
# This results in printing all pairs (i, j) where j <= i.
# Outer loop: i goes from 0 to 3
for i in range(4):
    # Inner loop: j goes from 0 to 3
    for j in range(4):
        # If j is greater than i, exit the inner loop
        if j > i:
            break
        # Print the current values of i and j
        print(i, j)
```

```python
# Example: Unpacking three values from an iterator of tuples
# The following code demonstrates unpacking values from a list of tuples.
# Each tuple in 'triplets' contains three elements.
# The for loop unpacks each tuple into variables a, b, and c in each iteration.
triplets = [(1, 'a', True), (2, 'b', False), (3, 'c', True)]
for a, b, c in triplets:
    print(f"a={a}, b={b}, c={c}")
```

```python
# This code demonstrates a while loop that halves the value of x in each iteration,
# adds it to the running total, and stops if the total exceeds 500.
# The loop starts with x = 256 and continues halving x (using integer division)
# until x becomes 0 or the total exceeds 500.

# Initialize x with 256
x = 256
# Initialize total with 0
total = 0
# Start the while loop; continue as long as x > 0
while x > 0:
    # If total exceeds 500, exit the loop
    if total > 500:
        break
    # Add current x to total
    total += x
    # Halve x using integer division
    x = x // 2
# Output the final total

print("x =", x, "total =", total)
```
